```python
### visualisasi data
```

```python
df = pd.DataFrame(data)
```

```python
# Plot Bar untuk NoChild
plt.figure(figsize=(10, 4))
plt.bar(df['MemberID'], df['NoChild'])
plt.xlabel('MemberID')
plt.ylabel('NoChild')
plt.title('Jumlah Anak per Anggota')
plt.show()
```

```python
# Plot Scatter untuk tenure_months dan avg_monetary
plt.figure(figsize=(10, 6))
plt.scatter(df['tenure_months'], df['avg_monetary'], c=df['NoChild'], cmap='viridis', alpha=0.7)
plt.colorbar(label='NoChild')
plt.xlabel('tenure_months (Bulan)')
plt.ylabel('avg_monetary')
plt.title('Hubungan antara tenure_months dan avg_monetary dengan Warna berdasarkan NoChild')
plt.show()
```

```python
# Box Plot untuk tenure_months
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 2)
plt.boxplot(df['tenure_months'])
plt.xlabel('tenure_months (Bulan)')
plt.title('Box Plot tenure_months')
```

```python
# Box Plot untuk avg_monetary
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 3)
plt.boxplot(df['avg_monetary'])
plt.xlabel('avg_monetary')
plt.title('Box Plot avg_monetary')
```

```python
# Box Plot untuk NoChild berdasarkan avg_monetary
plt.subplot(2, 2, 4)
plt.boxplot([df[df['NoChild'] == 1]['avg_monetary'], df[df['NoChild'] == 2]['avg_monetary']])
plt.xlabel('NoChild')
plt.xticks([1, 2], ['1 Child', '2+ Children'])
plt.title('Box Plot avg_monetary Berdasarkan NoChild')
```

```python
##analisis korelasi matriks
correlation_matrix = data.corr()
```

```python
import seaborn as sns
# Membuat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriks Korelasi")
plt.show()
```

```python
# Rumus SMA = (X1 + X2 + X3 + ... + Xn) / n
# Buat DataFrame
df = pd.DataFrame({'Data': data})

# Hitung SMA dengan periode 3
sma_period = 3
df['SMA'] = df['Data'].rolling(window=sma_period).mean()
print(df)
```
